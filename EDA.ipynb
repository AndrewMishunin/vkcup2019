```python
questions = pd.read_csv("vkcup/data.csv", sep=";", index_col="ID")
test = pd.read_csv("vkcup/test.csv", sep=";", index_col="ID")
test['Answer'] = 0
train = pd.read_csv("vkcup/train.csv", sep=";", index_col="ID")
data = questions.join(train)
drop_weird_rows(data)
```

```python
train=data[data.index<20000]
val=data[(data.index<25000) & (data.index>20000)]
```

```python
word_counter_analytics = WordCounterAnalytics(data)

def apply_word_counter_analytics(df):
    map_methods_from_class(df, word_counter_analytics, 'Question')
```

```python
for df in [train, val]:
    apply_word_counter_analytics(df)
    map_question_based_analytics(df)
    apply_postprocess_questionanalytics(df)
    apply_postprocess_final_analytics(df)
```
